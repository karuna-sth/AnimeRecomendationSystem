```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Anime.csv", index_col=False)
```

```python
anime = df[['Rank', 'Name', 'Tags', 'Studio', 'staff', 'Rating']]
```

```python
anime.head()
```

```python
anime['staff'][1]
```

```python
anime.isnull().sum()
```

```python
anime.dropna(inplace=True)
```

```python
anime.isnull().sum()
```

```python
anime.duplicated().sum()
```

```python
anime.shape
```

```python
list(filter(None, anime.iloc[1].Tags.split(",")))
```

```python
anime['Tags'] = [list(filter(None, x.split(","))) for x in anime['Tags']]
```

```python
anime.head()
```

```python
def Extdirector(staff):
  li = staff.split(",")
  for i in li:
    if "Director" in i:
      return (i.split(":")[0]).replace(" ", "")
  return 'Unknown'
```

```python
staff = anime.iloc[0].staff
Extdirector(staff)
```

```python
anime['Director'] = anime['staff'].apply(Extdirector)
```

```python
anime.head()
```

```python
del anime['staff']
```

```python
def tostr(value):
  value = " ".join(value)
  return value
```

```python
anime['tags'] = anime['Tags'].apply(tostr) + anime['Studio'] + anime['Director']
```

```python
anime.head()
```

```python
final_anime_data = anime[["Name", "Rating", 'tags']]
```

```python
final_anime_data.head()
```

```python
final_anime_data['tags'] = final_anime_data['tags'].apply(lambda x:x.lower())
```

```python
final_anime_data.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split(" "):
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('action adventure fantasy shounen demons')
```

```python
final_anime_data['tags'] = final_anime_data['tags'].apply(stem)
```

```python
final_anime_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2000)
```

```python
vectors = cv.fit_transform(final_anime_data['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend_anime(anime):
    anime_index = final_anime_data[final_anime_data['Name'] == anime].index[0]
    anime_list = sorted(list(enumerate(similarity[anime_index])), reverse=True,  key=lambda x:x[1])[1:10]
    for i in anime_list:
        print(final_anime_data.iloc[i[0]].Name)
```

```python
recommend_anime('Fullmetal Alchemist: Brotherhood')
```
